```python
import matplotlib.pyplot as plt
from numpy import *
```

```python
f_in = [50, 100, 500, 1000, 1600, 5000, 10000, 50000]
gain = [.006, .007, .163, .675, 1, .225, .036, .005]

gain = [20*log10(g) for g in gain]
# f_in = [log10(2*pi*f) for f in f_in]

fig, ax = plt.subplots()
ax.semilogx(f_in, gain)
ax.scatter(f_in, gain)
ax.set_xlim((0, 10**5))
ax.set_title('Gain of 3 Stage Amplifier')
ax.set_xlabel(r'$\omega$ [rad/s]')
ax.set_ylabel('dB')
```

```python
f_in = [200, 500, 1000, 1200, 1400, 1600, 1800, 2000, 2200, 2600, 3000, 3500, 4000, 5000, 6000, 7000, 8000, 9000, 10000]
vin = .500*0.01
gain = array([.250, 3, 12.9, 16.6, 19, 20.2, 20, 19.4, 18.4, 15.8, 12.8, 9.68, 7.44, 4.56, 2.9, 1.94, 1.34, .968, .720])/vin

assert len(f_in) == len(gain)

gain = [20*log10(g) for g in gain]
# f_in = [log10(2*pi*f) for f in f_in]

fig, ax = plt.subplots(figsize=(5,5))
ax.semilogx(f_in, gain)
ax.scatter(f_in, gain)
ax.set_xlim((10**2, 10**5))
ax.set_title('Gain of 3 Stage Amplifier')
ax.set_xlabel(r'$\omega$ [rad/s]')
ax.set_ylabel(r'$V_{out} \slash V_{in}$')
```

```python
integrand = [0]
integrand += [(f_in[i] - f_in[i-1])*gain[i]**2 for i in range(1, len(f_in))]

prev = 0
integral = [0]
for pt in integrand[1:]:
    integral.append(pt+prev)
    prev = pt

fig, ax = plt.subplots()
ax.scatter(f_in, gain )
# ax.scatter(f_in, integral)
ax.set_xlim((0, 10**4))
ax.set_title('Gain Squared of 3 Stage Amplifier')
ax.set_xlabel(r'f [Hz]')
ax.set_ylabel(r'$Av^2$')
```
