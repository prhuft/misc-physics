Eccentric Anomaly

```python
from numpy import *
import matplotlib.pyplot as plt
```

```python
eps = .75
a = 1
iters = 3 # correct to O(e**3), or 6 decimals
pts = 20

psi_m = lambda psi_n,theta: theta + eps*sin(psi_n)
r = lambda psi: a*(1 - cos(psi))
phi = lambda psi: arccos((cos(psi)-eps)/(1-eps*cos(psi)))

psi_arr = empty(pts)
r_arr = empty(pts)
phi_arr = empty(pts)
th_pts = linspace(0,2*pi,pts)
psi0 = 0

print("r        | psi      | phi      | t_step ")
for m,theta in enumerate(th_pts):
    psi = psi0
    for i in range(iters):
        psi = psi_m(psi,theta)
    psi_arr[m] = psi
    r_arr[m] = r(psi_arr[m])
    phi_arr[m] = phi(psi_arr[m])   
    
    print(f"{r_arr[m]:.6f} | {psi_arr[m]:.6f} | {phi_arr[m]:.6f} \
| {m} ")
```

```python
# for 20 pts
pts = [[a*(cos(psi) - eps),a*sqrt(1-eps**2)*sin(psi)] \
       for psi in psi_arr] 
pts = transpose(pts)
plt.plot(pts[0],pts[1])
```

```python
# for 100 pts
pts = [[a*(cos(psi) - eps),a*sqrt(1-eps**2)*sin(psi)] \
       for psi in psi_arr] 
pts = transpose(pts)
plt.plot(pts[0],pts[1])
```
