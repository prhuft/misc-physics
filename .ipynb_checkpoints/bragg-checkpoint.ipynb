```python
class Lattice:
    
    ux = array([1,0])
    uy = array([0,1])
    
    def __init__(self,gridx,basis=None):
        """
        gridx: the lattice width
        basis: a list of two-vector basis vectors
        """
        
        self.basis = basis
        self.v1 = self.basis[0]
        self.v2 = self.basis[1]
        self.gridx = gridx
        x0 = 0
        y0 = 0
        dx0 = dot(v1,ux)
        self.dx = dot(v2,ux)
        self.dy = dot(v1,uy)
        self.gridy = int(gridx*dx/dy)
    
    def plot(self, xlabel='x', ylabel='y', title='lattice',
             drawunit=False):
        """plot the lattice"""
        #TODO: add functionality to draw the unit cells
        xpts = array([x0 + j*dx for j in range(gridN)])

        fig, ax = plt.subplots()
        ax.set_aspect('equal')
        ax.set_title(title)
        ax.set_xlabel(xlabel)
        ax.set_ylabel(ylabel)
        ax.set_xlim((-dx, (gridx+1)*dx))
        ax.set_ylim((-dy, gridy*dy))
        for i in range(gridy):
            x = xpts + dx0*(i % 2)
            y = array([y0]*gridx)+dy*i
            ax.scatter(x, y, c='k')
            
    def rotate(self):
        """rotate the grid by theta via a transformation to the basis vectors"""
        # TODO: add code to do this
        
def get_grid(dx,dy,xnum,ynum):
    """
    Retunr xpts,ypts denoting centers of pts on a regular grid
    Args:
        dx,dy: center-center spacing in x,y directions
        xnum, ynum: number of columns, rows
    Return:
        xpts,ypts: lists of x coords,y coords for grid pts
    """
    xpts = []
    ypts = []
    for i in range(xnum):
        for j in range(ynum):
            ypts.append((1+i)*dy - dy*(1+ynum)/2)
            xpts.append((1+j)*dx - dx*(1+xnum)/2)
    return xpts, ypts

def figax(roi=None, xlabel=None, ylabel=None, aspect='equal', figkwargs={}):
    """
    return fig, ax. all params optional
    Args:
        roi: zoom to -roi, roi. None by default
        xlabel: None by default
        ylabel: None by default
        aspect: equal by default
    """
    fig,ax = plt.subplots(**figkwargs)
    if roi is not None:
        ax.set_xlim(-roi,roi)
        ax.set_ylim(-roi,roi)
    if aspect is not None:
        ax.set_aspect(aspect)
    if xlabel is not None:
        ax.set_xlabel(xlabel)
    if ylabel is not None:
        ax.set_ylabel(ylabel)
    return fig, ax
```

```python
## color objects

class RGB(tuple):
    """instantiate with a list [r,g,b]"""
    
    def __new__ (cls, iterable):
        return super(RGB, cls).__new__(cls, tuple(iterable))
    
    def __init__(self,iterable):
        self.minval = 1.0 if list(set([type(x) for x in iterable])) == [float] else 255
        self.type = type(self.minval)
    
    def __add__(self, value):
        return type(self)([self.type(min((x1+x2)/2, self.minval)) for x1,x2 in zip(self,value)])

    def __mul__(self, value):
        r,g,b = self
        return RGB([r*value, g*value, b*value])
    
class RGBA(RGB):
    """instantiate with a list [r,g,b,a]"""
    
    def __new__ (cls, iterable):
        return super(RGBA, cls).__new__(cls, tuple(iterable))
    
    def __mul__(self, value):
        r,g,b,a = self
        return RGBA([r, g, b, self.type(a*value)])
```

```python
#http://www.noah.org/wiki/Wavelength_to_RGB_in_Python
def wavelength_to_rgb(wavelength, gamma=0.8, linscale=1,alpha=None):
    '''
    Converts a given wavelength of light to an
    approximate RGB color value. 

    Based on code by Dan Bruton
    http://www.physics.sfasu.edu/astro/color/spectra.html
    
    Args:
        wavelength: int in range 380 to 750
        gamma=0.8: power scaling exponent
        linscale=1: linear scaling factor
        alpha=None: if a number between 0 and 1, the returned value
        will be a four-tuple (r,g,b,a)
    Return:
        rgb: a tuple-like RGB object (r,g,b) where r,g,b
        are in the range 0 to 255. If alpha set, returns (r,g,b,a)
    '''

    wavelength = float(wavelength)
    if wavelength >= 380 and wavelength <= 440:
        attenuation = 0.3 + 0.7 * (wavelength - 380) / (440 - 380)
        R = ((-(wavelength - 440) / (440 - 380)) * attenuation) ** gamma
        G = 0.0
        B = (1.0 * attenuation) ** gamma
    elif wavelength >= 440 and wavelength <= 490:
        R = 0.0
        G = ((wavelength - 440) / (490 - 440)) ** gamma
        B = 1.0
    elif wavelength >= 490 and wavelength <= 510:
        R = 0.0
        G = 1.0
        B = (-(wavelength - 510) / (510 - 490)) ** gamma
    elif wavelength >= 510 and wavelength <= 580:
        R = ((wavelength - 510) / (580 - 510)) ** gamma
        G = 1.0
        B = 0.0
    elif wavelength >= 580 and wavelength <= 645:
        R = 1.0
        G = (-(wavelength - 645) / (645 - 580)) ** gamma
        B = 0.0
    elif wavelength >= 645 and wavelength <= 750:
        attenuation = 0.3 + 0.7 * (750 - wavelength) / (750 - 645)
        R = (1.0 * attenuation) ** gamma
        G = 0.0
        B = 0.0
    else:
        R = 0.0
        G = 0.0
        B = 0.0
    R *= 255*linscale
    G *= 255*linscale
    B *= 255*linscale
    if alpha is not None:
        return RGB([int(R), int(G), int(B)]) # uses my tuple extension for nice adding - PH
    else:
        assert 0 <= alpha <= 255, "alpha must be in range [0,255]"
        return RGB([int(R), int(G), int(B), ceiling(alpha)])
```

```python
## simple Bragg diffraction over 1D bilayer. 

# assume interference is from rays that scatter from the same x location
lambda0 = 500
k = 2*pi/lambda0
thetax = pi/2
thetay = pi/2

# compute interference at each position in x, at top surface
xnum = 100 # units are kind of arbitrary
ynum = xnum
field0 = zeros((xnum,ynum))
rng = default_rng()
d0 = 200
dsigma = lambda0/100
for i in range(xnum):
    for j in range(ynum):
        # assume field is a cosine with phase = 0 at top plane
        dx = rng.normal(loc=d0,scale=dsigma)
        dy = rng.normal(loc=d0,scale=dsigma)
        field0[i,j] += cos(k*2*dx*sin(thetax)) + cos(k*2*dy*sin(thetay))
plt.title(f"Surface variation {floor(dsigma)} nm")
plt.imshow(field0**2)
```

```python
## simple Bragg diffraction over 1D bilayer. 

# now consider light that reflects within a certain veiwing angle

# assume interference is from rays that scatter from the same x location
theta_range = linspace(-pi/6,pi/6,20)

# compute interference at each position in x, at top surface
xnum = 100 # units are kind of arbitrary
ynum = xnum
field0 = zeros((xnum,ynum))
seed = 12345
rng = default_rng(seed)
d0 = 200
dsigma = d0/10

for i in range(xnum):
    for j in range(ynum):
        
        # assume field is a cosine with phase = 0 at top plane
        dx = rng.normal(loc=d0,scale=dsigma)
        dy = rng.normal(loc=d0,scale=dsigma)
        for thx in theta_range:
            for thy in theta_range:
                field0[i,j] += cos(k*2*dx*sin(thx)) + cos(k*2*dy*sin(thy))
plt.title(f"Surface variation {floor(dsigma)} nm")
plt.imshow(field0**2)
```

```python
## less simple Bragg diffraction over 1D bilayer. 

# now consider light that reflects within a certain veiwing angle
# and is polychromatic

theta_range = linspace(-pi/6,pi/6,20)

# compute interference at each position in x, at top surface
xnum = 10 # units are kind of arbitrary
ynum = xnum
seed = 12345
rng = default_rng(seed)
d0 = 200
dsigma = d0/10

# choose wavelengths >= 1nm apart; assume don't interfere
lambda_list = linspace(400,700,1)
# lambda_weight = lambda x: exp(-(x-mean(lambda_list))**2/(2*std(lambda_list)**2))

rgb_list = [[RGB([0,0,0]) for i in range(xnum)]
           for j in range(ynum)]

norm = 2*xnum*ynum*len(theta_range)**2

for step,lmbda in enumerate(lambda_list):

    k = 2*pi/lmbda
    field0 = zeros((xnum,ynum))
    
    t0 = time()
    for i in range(xnum):
        for j in range(ynum):

            # assume field is a cosine with phase = 0 at top plane
            dx = rng.normal(loc=d0,scale=dsigma)
            dy = rng.normal(loc=d0,scale=dsigma)
            for thx in theta_range:
                for thy in theta_range:
                    field0[i,j] += cos(k*2*dx*sin(thx)) + cos(k*2*dy*sin(thy))
                    
    field0 = field0/amax(field0)
    for i in range(xnum):
        for j in range(ynum):
            rgb_list[i][j] += wavelength_to_rgb(lmbda, linscale=field0[i,j]**2)
                    
    print(f"completed step {step} in {time()-t0} s")
    
plt.title(f"Surface variation {floor(dsigma)} nm")
plt.imshow(rgb_list)
# fname = 'bragg_plot_{xnum}x{ynum}_d{int(d0)}_dmu{int(dmu)}.png'
# plt.savefig(fname, bbox_inches='tight')
```

```python
# surprisingly difficult to get plot to show variation in intensity of 
# a color. there should be some regions that reflect one wavelength more
# than another, but adding rgb's from wavelength does depend on relative
# amplitudes of two waves. using the transparency parameter does not 
# resolve the issue. should test simple cases of adding two colors
# and tune to get desired result in just one square, then implement
# in the Bragg simulation.

## less simple Bragg diffraction over 1D bilayer. 

# now consider light that reflects within a certain veiwing angle
# and is polychromatic

theta_range = linspace(-0.3,0.3,10)

# compute interference at each position in x, at top surface
xnum = 10 # units are kind of arbitrary
ynum = xnum
seed = 12345
rng = default_rng(seed)
d0 = 200
dsigma = d0

# choose wavelengths >= 1nm apart; assume don't interfere
lambda_list = linspace(400,500,10)
# lambda_weight = lambda x: exp(-(x-mean(lambda_list))**2/(2*std(lambda_list)**2))

rgba_list = [[RGBA([0,0,0,255]) for i in range(xnum)]
           for j in range(ynum)]

norm = 2*xnum*ynum*len(theta_range)**2

for step,lmbda in enumerate(lambda_list):

    k = 2*pi/lmbda
    field0 = zeros((xnum,ynum))
    
    t0 = time()
    for i in range(xnum):
        for j in range(ynum):

            # assume field is a cosine with phase = 0 at top plane
            dx = rng.normal(loc=d0,scale=dsigma)
            dy = rng.normal(loc=d0,scale=dsigma)
            for thx in theta_range:
                for thy in theta_range:
                    field0[i,j] += cos(k*2*dx*sin(thx)) + cos(k*2*dy*sin(thy))
                    
    int0 = field0**2
    int0 /= amax(int0)
    
    for i in range(xnum):
        for j in range(ynum):
            rgba_list[i][j] += wavelength_to_rgb(lmbda, alpha=255*(1-int0[i,j]))
                    
    print(f"completed step {step} in {time()-t0} s")

plt.title(f"Surface variation {floor(dsigma)} nm")
plt.imshow(rgba_list)
# fname = 'bragg_plot_{xnum}x{ynum}_d{int(d0)}_dmu{int(dmu)}.png'
# plt.savefig(fname, bbox_inches='tight')
```
